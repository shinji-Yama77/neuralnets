```python
import numpy as np
from neuralnetwork import neuralnetwork 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('healthcare.csv')




```

```python
df.head(10)
```

```python
X = df.drop(columns=['Id', 'Outcome'])
```

```python
X.head(10)
```

```python
input_variables_array = np.array(df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']].values).T
```

```python
train_samples = int(0.7 * input_variables_array.shape[1])
train_samples
```

```python
targets_array = np.array([df['Outcome'].values])
```

```python
X_train = input_variables_array[:, :train_samples]
y_train = targets_array[:, :train_samples]
```

```python
X_test = input_variables_array[:, train_samples:]
X_test.shape
y_test = targets_array[:, train_samples:]
y_test.shape
```

```python
bb = neuralnetwork(8, 100, 1)

bb.train(X_train, y_train, 3000)
```

```python
bb.binary_loss
```

```python
predictions = bb.feed_forward(X_test)
```

```python
binary_predictions = (predictions >= 0.5).astype(int)
```

```python
binary_predictions
```

```python

print('Testing Accuracy: %f'%(np.mean(binary_predictions == y_test) * 100))
```
