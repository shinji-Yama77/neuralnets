```python
input_variables_array = np.array(df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']].values).T
```

```python
nn = neuralnetwork(8, 50, 1)
```

```python
nn.train(input_variables_array, targets_array, 100)

nn.binary_loss
```

```python
predictions = nn.feed_forward(input_variables_array)
```

```python
predictions_df = pd.DataFrame({'Predictions': predictions[0]})
```
